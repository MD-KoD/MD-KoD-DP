```python
def clean_mct_type(value):
    # 값이 숫자형(float 또는 int)이거나, 숫자로 변환 가능한 문자열인 경우 '기타' 처리
    try:
        float(value)  # 숫자형이거나 숫자로 바뀌는 문자열이면
        return '기타'
    except:
        return value  # 나머지는 그대로

# 적용
info_df['MCT_TYPE_CLEAN'] = info_df['MCT_TYPE'].apply(clean_mct_type)
```

```python
# 예시 MCT_TYPE Series
mct_series = info_df['MCT_TYPE']

# 카테고리 매핑 딕셔너리
category_map = {
    '한식': ['단품요리 전문', '가정식', '분식', '떡/한과', '기사식당', '야식'],
    '일식': ['일식', '꼬치구이'],
    '중식': ['중식'],
    '양식': ['양식', '피자', '스테이크', '패밀리 레스토랑'],
    '아시아기타': ['동남아/인도음식', '기타세계요리'],
    '패스트푸드': ['햄버거', '샌드위치/토스트', '도너츠', '도시락'],
    '디저트/카페': ['커피', '차', '아이스크림/빙수', '주스', '베이커리'],
    '주점/주류': ['맥주/요리주점', '민속주점', '포장마차'],
    '기타': ['구내식당/푸드코트', '부페']
}

# mapping function 정의
def map_category(x):
    for category, keywords in category_map.items():
        if x in keywords:
            return category
    return '기타'  # 매핑되지 않는 값은 기타로 처리

# 대분류 컬럼 생성
info_df['MCT_TYPE_GROUP'] = mct_series.apply(map_category)

# 원-핫 인코딩 수행
mct_onehot = pd.get_dummies(info_df['MCT_TYPE_GROUP'], prefix='MCT_TYPE_GROUP')

# 기존 DataFrame에 병합
info_df = pd.concat([info_df, mct_onehot], axis=1)

# 중복된 열 제거
info_df = info_df.loc[:, ~info_df.columns.duplicated()]
# 결과 예시 보기
print(info_df[['MCT_TYPE', 'MCT_TYPE_GROUP']])
```

```python
# 기존 원핫 컬럼 제거 (혹은 새로 DataFrame 구성)
info_df = info_df.drop(columns=[col for col in info_df.columns if 'MCT_TYPE_GROUP_' in col])

# 다시 원핫 인코딩
one_hot = pd.get_dummies(info_df['MCT_TYPE_GROUP'], prefix='MCT_TYPE_GROUP')
info_df = pd.concat([info_df, one_hot], axis=1)
```
